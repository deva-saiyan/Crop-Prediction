# Split Train and Test Data

```python
x = df.drop(columns='crop_name' , axis=1)
 
y = df['crop_name']
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y ,test_size=0.25 ,random_state=15 )
```

```python
print(f'x {x.shape}\ny {y.shape}')
```

# Model Building

Linear Model

```python
from sklearn.svm import SVC

model_0 = SVC()

model_0.fit(x_train , y_train)

svmc = model_0.predict(x_test)

```

DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


dt = DecisionTreeClassifier()

dt.fit(x_train, y_train)

dtc = dt.predict(x_test)
print("Optimized DT Accuracy:", accuracy_score(y_test, dtc))
```

Model 1 - AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier

model_1 = AdaBoostClassifier()

model_1.fit(x_train , y_train)

AdaBoost = model_1.predict(x_test)
```

Model 2 - GradientBoost Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

model_2 = GradientBoostingClassifier()
model_2.fit(x_train, y_train)

GBC = model_2.predict(x_test)
```

Model 3 - XGBoost Classification

```python
from xgboost import XGBClassifier
model_3 = XGBClassifier()

model_3.fit(x_train , y_train)

XGC = model_3.predict(x_test)
```

Accuracy in Classification Models

```python
from sklearn.metrics import accuracy_score

models  = [svmc ,dtc,AdaBoost, GBC,  XGC,]
names   = ['svmc','dtc',"AdaBoost", "GBC",  "XGC"]

for i, j in zip(models, names):
    accuracy = accuracy_score(y_test, i)
    print(f"{j} Accuracy: {accuracy * 100}%")
```

Decision Tree + Gradient Boost

```python
from xgboost import XGBClassifier

model_5 = XGBClassifier(
    max_depth=5,
    learning_rate=0.1,
    n_estimators=200
)

model_5.fit(x_train, y_train)
XGBC_dt = model_5.predict(x_test)



```

# SVM + Boosting

```python
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np

# Step 1: Train XGBoost (feature generator)
xgb = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8
)
xgb.fit(x_train, y_train)

# Step 2: Generate boosted leaf embeddings
train_features = xgb.apply(x_train)
test_features = xgb.apply(x_test)

# Convert to 2D
train_features = train_features.reshape(train_features.shape[0], -1)
test_features = test_features.reshape(test_features.shape[0], -1)

# Step 3: Train SVM on boosted features
svm = SVC(kernel='rbf', C=10, gamma='scale')
svm.fit(train_features, y_train)

# Step 4: Prediction
XGB_svm = svm.predict(test_features)

# Step 5: Accuracy
acc = accuracy_score(y_test, XGB_svm)
print("XGBoost + SVM Accuracy:", acc * 100)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Models
svm = SVC(probability=True, kernel='rbf')
gb = GradientBoostingClassifier()

# Train
svm.fit(x_train, y_train)
gb.fit(x_train, y_train)

# Probabilities
p1 = svm.predict_proba(x_test)
p2 = gb.predict_proba(x_test)

# Average ensemble
GB_svm = np.argmax((p1 + p2) / 2, axis=1)

# Accuracy
print("Accuracy:", accuracy_score(y_test, GB_svm))
```
